```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from datetime import datetime, timedelta
```

```python
def load_and_describe(file_name):
    try:
        """Load The Data Set For Better Understanding"""
        data=pd.read_csv(file_name)
        print(f" Data Set {file_name} is Loaded Successfully\n")
        print(f"Dataset Head:\n")
        display(data.head())
        print(f"\n Basic Information\n")
        display(data.info())
        print(f"\n Data Statistics\n")
        display(data.describe())
        print("\n Skewness\n")  #
        print(data.skew(numeric_only=True))
        print(f"\n Kurtosis \n")
        display(data.kurtosis(numeric_only=True))
        return data
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
data=load_and_describe("Indian Airlines.csv")
```

```python
today = datetime(2025, 6, 7)

# Step 1: Departure date
data['departure_date'] = data['days_left'].apply(lambda x: today + timedelta(days=int(x)))
data['day'] = data['departure_date'].dt.day
data['month'] = data['departure_date'].dt.month

# Step 2: Map departure_time and arrival_time to numeric hour values
time_mapping = {
    'Early_Morning': 5,
    'Morning': 9,
    'Afternoon': 13,
    'Evening': 18,
    'Night': 22,
    'Late_Night': 2
}

data['departure_time_numeric'] = data['departure_time'].map(time_mapping)
data['arrival_time_numeric'] = data['arrival_time'].map(time_mapping)

# Step 3: Ensure duration is numeric
# If it's already a float like 2.17, this will just ensure it's numeric
data['duration_numeric'] = pd.to_numeric(data['duration'], errors='coerce')

# Check the result
data.head()
```

```python
# Droping the Useless Column
data = data.drop(columns=['Unnamed: 0'])
```
